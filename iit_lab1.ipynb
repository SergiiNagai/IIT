__Завдання 2.__ \
Розв’язати систему алгебраїчних рівнянь за допомогою формул Крамера і виконати перевірку за допомогою: \
a) матричного множення; \
b) оберненої матриці; \
c) функції numpy.linalg.solve(). \
Порівняти всі рішення за допомогою функції numpy.allclose() \
\begin{cases} 2x_1 + x_3 + 4x_4 = 9 \\ x_1 + 2x_2 + x_3 + x_4 = 8 \\ 2x_1 + x_2 + x_3 + x_4 = 5 \\ x_1 + x_2 + 2x_3 + x_4 = -1 \end{cases}

```python
import numpy as np

def rozwiaz_system(X, y):
    # 1:Знаходимо головний визначник матриці X
    vuzn_golovnii = np.linalg.det(X)

    # 2:Ініціалізуємо список для збереження розв'язків
    rozwiazania = []

    # 3:Розв'язуємо систему рівнянь за допомогою формул Крамера
    for i in range(X.shape[1]):
        # 4:Замінюємо стовпець i матриці X на вектор y
        ai = X.copy()
        ai[:, i] = y

        # 5:Знаходимо визначник нової матриці
        vuzn_i = np.linalg.det(ai)

        # 6:Знаходимо значення невідомої змінної xi
        xi = vuzn_i / vuzn_golovnii

        # 7:Додаємо xi до списку розв'язків
        rozwiazania.append(xi)

    # 8:Перевіряємо отримані значення за допомогою матричного множення
    ax = np.dot(X, rozwiazania)
    if np.allclose(ax, y):
        print("Перевірка за допомогою  матричного множення пройдена.")
    else:
        print("Перевірка за допомогою  матричного множення не пройдена.")

    return rozwiazania

# 9:Вхідні дані
X = np.array([[2, 0, 1, 4],
              [1, 2, -1, 1],
              [2, 1, 1, 1],
              [1, -1, 2, 1]])
y = np.array([9, 8, 5, -1])

# 10:Розв'язуємо систему рівнянь
rozwiazania = rozwiaz_system(X, y)
print("Розв'язок системи алгебраїчних рівнянь за допомогою формул Крамера: ")
print(rozwiazania)

# 11:Перевірка за допомогою оберненої матриці
inverse_k = np.linalg.inv(X)
im_check = np.dot(inverse_k, y)
print("Перевірка за допомогою оберненої матриці:", im_check)

# 12:Перевірка за допомогою функції numpy.linalg.solve()
f_check = np.linalg.solve(X, y)
print("Перевірка за допомогою функції numpy.linalg.solve():", f_check)

# 13:Порівняння з результатом перевірки за допомогою оберненої матриці
is_equal = np.allclose(rozwiazania, im_check)
print("Порівняння з результатом перевірки за допомогою оберненої матриці:", is_equal)

# 14:Порівняння з результатом перевірки за допомогою функції numpy.linalg.solve()
is_equal = np.allclose(rozwiazania, f_check)
print("Порівняння з результатом перевірки за допомогою функції numpy.linalg.solve():", is_equal)
```

Обчислити значення матричного виразу \
а) з використанням універсальних функцій бібліотеки NumPy. \
b) за допомогою ітеративних конструкцій (з використанням циклів, спискових включень тощо). \
с) для обох випадків порахувати час виконання скрипту та зробити висновок. \
Порівняти всі рішення за допомогою функції numpy.allclose() \
$
2(2A + B) - B(A - B), де A = \begin{bmatrix}
2 & 3 & 1\\\
4 & -1 & 0\\\
0 & 1 & 2
\end{bmatrix}, B = \begin{bmatrix}
9 & 8 & 7\\\
2 & 7 & 3\\\
4 & 3 & 5
\end{bmatrix}

```python
import numpy as np
import time

# 1:Обчислюємо значення матричного виразу з використанням універсальних функцій бібліотеки NumPy
def task_num_py(a, b):
    return a @ (2 * a + b) - b @ (a - b)

# 2:Обчислюємо значення матричного виразу за допомогою ітеративних конструкцій та виводимо результат
def task_iterative(a, b):
    print("\nРезультати обрахунків ітеративною функцією:")
    c = []
    for i in range(len(a)):
        row = []
        for j in range(len(a[0])):
            row.append(a[i][j] - b[i][j])
        c.append(row)
    print("\nA - B: ", c)

    d = []
    for i in range(len(b)):
        row = []
        for j in range(len(c[0])):
            s = 0
            for k in range(len(c[0])):
                s += b[i][k] * c[k][j]
            row.append(s)
        d.append(row)
    print("\nB * (A - B): ", d)

    e = []
    for i in range(len(a)):
        row = []
        for j in range(len(a[0])):
            row.append(a[i][j] * 2)
        e.append(row)
    print("\n2 * A: ", e)

    f = []
    for i in range(len(a)):
        row = []
        for j in range(len(a[0])):
            row.append(e[i][j] + b[i][j])
        f.append(row)
    print("\n(2 * A + B): ", f)

    h = []
    for i in range(len(a)):
        row = []
        for j in range(len(a[0])):
            s = 0
            for k in range(len(a[0])):
                s += a[i][k] * f[k][j]
            row.append(s)
        h.append(row)
    print("\nA * (2 * A + B): ", h)

    z = []
    for i in range(len(h)):
        row = []
        for j in range(len(h[0])):
            row.append(h[i][j] - d[i][j])
        z.append(row)
    print("\nКінцевий результат A * (2 * A + B) - B * (A - B): ", z)
    return z

# 3:Вводимо матричні вирази згідно варіанту
A = np.array([[2, 3, 1], [4, -1, 0], [0, 1, 2]])
B = np.array([[9, 8, 7], [2, 7, 3], [4, 3, 5]])

# 4:Виклик та обрахунок часу виконання функції NumPy
start_time = time.perf_counter()
numpyResult = task_num_py(A, B)
end_time = time.perf_counter()
numpy_time = end_time - start_time

# 5::Виклик та обрахунок часу виконання ітеративної функції
start_time = time.perf_counter()
iterativeResult = task_iterative(A, B)
end_time = time.perf_counter()
iterative_time = end_time - start_time

print("\nРезультати обрахунків функцією Numpy:")
print("\nA + B:\n", (A + B))
print("\nA - B:\n", (A - B))
print("\nA * (2 * A + B):\n", A @ (2 * A + B))
print("\nB * (A - B):\n", B @ (A - B))
print("\nКінцевий результат - A * (2 * A + B) - B * (A - B):\n", A @ (2 * A + B) - B @ (A - B))

comparison = np.allclose(numpyResult, iterativeResult)
print(f"\nРезультати обрахунків обома методами співпадають: {comparison}")

print(f"\nЧас виконання обрахунків функціями NumPy: {numpy_time:.6f} секунд")
print(f"Час виконання обрахунків ітеративними функціями: {iterative_time:.6f} секунд")
```
